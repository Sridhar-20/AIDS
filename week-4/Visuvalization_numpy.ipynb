```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
# df.head(10)
# df.info()
# df.describe()
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.isnull().sum()
```

```python
df.head(10)
```

```python
# how does passenger class relate to survival rates
# plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Pclass', y='Survived', errorbar=None, palette=['red', 'green', 'skyblue'],edgecolor='black')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# id there a relationship b/w passenger's age vs.fare
sns.scatterplot(data=df,x='Age',y='Fare',palette='Greys',hue='Survived',alpha=0.6)
plt.title('Relationship B/w Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# pairwise relationship b/w age,fare,and passenger class 
dat = df[['Age','Fare','Pclass']].copy()
sns.pairplot(data=dat,diag_kind='kde')
plt.suptitle('Relationship Between Age, Fare, and Passenger Class', y=1.02)
plt.show()
```

```python
# correlation b/w  different numerical features sets in dataset
numeric_cols = df.select_dtypes(include='number')
corr_matrix = numeric_cols.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',linewidth=0.9)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Copy vs View
# copy
arr1 = np.array([1, 2, 3, 4])
arr2 = arr1.copy()  # Create a copy of arr1
arr2[0] = 10  # Modify the copy
print("Original Array:", arr1)  # [1, 2, 3, 4]
print("Copied Array:", arr2)    # [10, 2, 3, 4]
# view
arr1 = np.array([1, 2, 3, 4])
arr2 = arr1.view()  # Create a view of arr1
arr2[0] = 10  # Modify the view
print("Original Array:", arr1)  # [10, 2, 3, 4]
print("View Array:", arr2)      # [10, 2, 3, 4]
```
