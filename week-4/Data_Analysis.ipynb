```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[2,4,6,8,10]
})
df.plot(xlabel='x-axis',ylabel='y-axis',color='blue',title='Line Plot',marker='*',linestyle='--',markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'Months': ['jan', 'feb', 'mar', 'apr', 'may', 'jun'],
    'Savings_amt': [4500, 1500, 4000, 2000, 2500, 5000]
})
df.plot(x='Months', y='Savings_amt', color='green', marker='*', linestyle='--', markersize=10)
plt.xlabel('Months')
plt.ylabel('Savings_amt')
plt.title('Growth of Savings Account')
plt.show()
```

```python
# bar pLot
categories=['A','B','C','D','E']
values=[10,20,7,12,9]
plt.bar(categories,values,color=['red','yellow','orange','green','skyblue'],edgecolor='black',linewidth=3)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
ages=[20,22,24,24,25,26,27,27,28,30,31,32,34,35,35,37,38,38,39,40,40,41,43,45,47,49,50,50,51,51,52,54,55,56,57,56,58,58,59]
b=[20,30,40,50,60]
plt.hist(ages,bins=b,edgecolor='black',color='skyblue',linewidth=2)
plt.title('Histogram')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# PyChart
lang=['python','java','c','c++','javascript']
lang_per=[50,40,10,40,60]
my_colors = ['#E40303', '#FF8C00', '#FFED00', '#008026', '#004CFF']
plt.pie(lang_per,labels=lang,autopct='%1.1f%%',shadow=True,colors=my_colors)
plt.title('Pie Chart')
plt.show()
```

```python
# Scatter PLot
x=[5,10,15,20,25,32,34,35,35,37,38,38,39,40]
y=[10,20,25,30,35,52,54,55,56,57,56,58,58,59]
plt.scatter(x,y,color='green',s=160,edgecolor='black',marker='*')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
